```python
A.isna().sum()  ### all are zero mens no missing values
```

```python
#Changing each data into numeric
for i in A.columns:
    if i=='pdate':
        continue
    else:
        A[i]=pd.to_numeric(A[i],errors='coerce')
```

```python
A.head()
```

 On changing the value to Numeric all the values containing str was filled with NaN

#contained Missing values then it will be replace by the median of the training data

# RANDOM FOREST CLASSIFICATION

```python
X=A[['aon', 'daily_decr30', 'daily_decr90', 'rental30', 'rental90',
       'last_rech_date_ma', 'last_rech_date_da', 'last_rech_amt_ma',
       'cnt_ma_rech30', 'fr_ma_rech30', 'sumamnt_ma_rech30',
       'medianamnt_ma_rech30', 'medianmarechprebal30', 'cnt_ma_rech90',
       'fr_ma_rech90', 'sumamnt_ma_rech90', 'medianamnt_ma_rech90',
       'medianmarechprebal90', 'cnt_da_rech30', 'fr_da_rech30',
       'cnt_da_rech90', 'fr_da_rech90', 'cnt_loans30', 'amnt_loans30',
       'maxamnt_loans30', 'medianamnt_loans30', 'cnt_loans90', 'amnt_loans90',
       'maxamnt_loans90', 'medianamnt_loans90', 'payback30', 'payback90']]
Y=A[['label']]

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X=X.apply(le.fit_transform)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=30)

from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(min_samples_leaf=9)
model=rfc.fit(xtrain,ytrain)
pred=model.predict(xtest)

from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,pred))
```

```python
import matplotlib.pyplot as plt
f,ax=plt.subplots(figsize=(10,12))
plt.barh(X.columns,model.feature_importances_)
```
